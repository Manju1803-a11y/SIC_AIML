```python
import sklearn

from sklearn.datasets import load_iris
```

```python

# Loading the iris dataset
iris = load_iris()
X_iris = iris.data
y_iris = iris.target
```

```python
X_iris, y_iris
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

Scale the features data

```python
scaler = StandardScaler()
```

```python
# Standardizing the data
scaled_iris = scaler.fit_transform(X_iris)
```

```python
# scaled_iris
```

```python
# Applying PCA
pca_iris = PCA()
pca_iris.fit(scaled_iris)
```

```python
# Explained variance ratio
explained_variance_iris = pca_iris.explained_variance_ratio_
```

```python
explained_variance_iris
```

```python
print(f"Explained variance by each component (iris): {explained_variance_iris}")
```

```python
pca_iris.components_
```

```python
import pandas as pd
```

```python
# Showing the components (loadings)
components_iris = pd.DataFrame(pca_iris.components_,
                               columns=iris.feature_names)
```

```python
print("Principal components (loadings) for iris:")
print(components_iris)
```

Classification

```python
from sklearn.ensemble import RandomForestClassifier

# Fitting the Random Forest model
rf_classifier = RandomForestClassifier()
```

```python
rf_classifier.fit(X_iris, y_iris)
```

```python
# rf_classifier.feature_importances_
```

```python
# Getting feature importances
importances_iris = rf_classifier.feature_importances_
```

```python
# Creating a DataFrame for feature importances
feature_importance_iris_df = pd.DataFrame({
    'Feature': iris.feature_names,
    'Importance': importances_iris
}).sort_values(by='Importance', ascending=False)
```

```python
print("Feature importances for iris:")
print(feature_importance_iris_df)
```

Regression

```python
from sklearn.datasets import fetch_california_housing
```

```python
# Loading the California Housing dataset
california_housing = fetch_california_housing()

X = california_housing.data
y = california_housing.target
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
 #Standardizing the data
scaled_X = scaler.fit_transform(X)

# Applying PCA
pca = PCA()
pca.fit(scaled_X)

# Explained variance ratio
explained_variance_california = pca.explained_variance_ratio_

# Showing the components (loadings)
components = pd.DataFrame(pca.components_,
                          columns=california_housing.feature_names)

print("Explained variance by each component (California Housing): ",
      explained_variance_california)
print("Principal components (loadings) for California Housing:",components)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Fitting the Random Forest model
rf = RandomForestRegressor()
rf.fit(X, y)

# Getting feature importances
importances = rf.feature_importances_

# Creating a DataFrame for feature importances
feature_importance = pd.DataFrame({
    'Feature': california_housing.feature_names,
    'Importance': importances}).sort_values(by='Importance', ascending=False)

print("Feature importances for California Housing:")
print(feature_importance)
```
